```python
def converter(string, nums):
    print(type(string), type(nums))
    
    if type(string) == str:
        string = string.split(",")
        numList01 = [int(strs) for strs in string]
        return numList01

    elif type(nums) == tuple:
        numList02 = list(nums)
        return numList02    

converter('1,2,3,4,5,6', (1,2,3,4,5,6))        
```

### 딕셔너리
* {key_1 : value_1, key_2:value_2}
* 딕셔너리 구조는 JSON 파일과 같다.
* JSON 확장자는 딕셔너리 구조이다. 

```python
my_dict = {
    "name": "John",
    "age": 25,
    "city": "New York",
    "gender": "Male",
    "occupation": "Engineer"
}
# print(list(my_dict.keys())[0])
# print(list(my_dict.values())[0])
print(list(my_dict.items())[0])
print(sorted(my_dict.items(), reverse=True))

```

```python
import operator
### operator 모듈 사용하기
    ### 계산을 도와주는 다양한 명령어 포함
print(sorted(my_dict.keys(), reverse = True, key = operator.itemgetter(0)))
# itemgetter()
    # 값이 0이면: key를 기준으로 정렬
    # 값이 1이면: value를 기준으로 정렬
```

```python
def priceInfo(indexNum):
    user = input()    
    indexNum = title.index(indexNum)
    totalList = [int(row[indexNum]) for row in data]
    print("최대:{}원,최소:{}원,합계:{}원, 평균:{}원".format(max(totalList), min(totalList), sum(totalList), round(sum(totalList) / len(totalList),2) ))
    
    if user == '최대':
        return max(totalList)
    if user == '최소':
        return min(totalList)
    if user == '합계':
        return sum(totalList)
    if user == '평균':
        return round(sum(totalList) / len(totalList), 2)
    
indexNum = '이용금액'
priceInfo(indexNum)
```
