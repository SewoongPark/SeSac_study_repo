```python
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud

texts = '''Thank you for joining us today. In the face of challenges, 
we find opportunities for growth and unity. 
Together, we can overcome any obstacle and build a brighter future. 
Let's embrace diversity, foster understanding, 
and work hand in hand towards a world of compassion and progress.'''

speechText = texts.replace("\n", "")

nltk.download('stopwords')

stopwords = set(stopwords.words('english'))
splitText = speechText.split(" ")
filters = [word for word in splitText if word.lower() not in stopwords]
filtered_text = " ".join(filters)
word_frequency = Counter(filters)
top_words = word_frequency.most_common(5)

# WordCloud 객체 생성
wordcloud = WordCloud(width=400, height=400, background_color='white').generate(filtered_text)

# 시각화
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
