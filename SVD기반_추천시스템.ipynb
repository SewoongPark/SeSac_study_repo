```python
data2['rating'] = data2['rating'].astype(float) # 연산을 위한 타입 변환
```

```python
user_ids = data2['user_id'].unique().tolist()
#print(len(user_ids))
user2idx_dict = { u : i for i, u in enumerate(user_ids)}
#print(user2idx_dict)
idx2user_dict = {i : u for i, u in enumerate(user_ids)}
#print(idx2user_dict)
```

```python
user_ids = data2['user_id'].unique().tolist() #user id 목록을 리스트에 담은 배열

user2idx_dict = {u : i for i, u in enumerate(user_ids)} #user_id에 index번호 메김
idx2user_dict = { i : u for i, u in enumerate(user_ids)} #u = user_id, i = index

book_ids = data2['ISBN'].unique().tolist() #isbn 목록을 리스트에 담은 배열

book2idx_dict = {b: i for i, b in enumerate(book_ids)} # b = index, i = isbn
idx2book_dict = {i: b for i, b in enumerate(book_ids)} # i = index, b = isbn

data2['user_idx'] = data2['user_id'].map(user2idx_dict)
'''
 data 데이터 프레임 내의 isbn이 dict의 b값과 일치하면 dict의 i값으로 변경,
 isbn -> index 한 값을 새로운 col에 추가
'''
data2['book_idx'] = data2['ISBN'].map(book2idx_dict)

num_users = data2['user_id'].nunique()
num_books = data2['ISBN'].nunique()
# 행렬분해를 위한 범위(가로, 세로)지정
print(num_users, num_books)

data2 = data2.sort_values(['user_idx', 'book_idx'])
data2 = data2.reset_index(drop = True)
data2
```

```python
data2['ISBN'].iloc[:len(data2)]
```

```python
from scipy.sparse import coo_matrix, csr_matrix, csc_matrix, linalg
sparse_arr = coo_matrix((data2['rating'],
 (data2['user_idx'],
  data2['book_idx'])),
  shape = (num_users, num_books))
sparse_arr
```

```python
print(sparse_arr.toarray())
```

```python
U, S, V = linalg.svds(sparse_arr, k = 1550) # 실험끝에 찾은 임계 값
```

```python
U.shape, S.shape, V.shape
```

```python
U, V, S
```

```python
test_user_id = 88705
test_user_df = data2[data2['user_id'] == test_user_id]
test_user_idx = test_user_df['user_idx'].iloc[0]
test_user_idx
```

```python
test_user_df
```

```python
pred_target_user = U[test_user_idx] @ np.diag(S) @ V
```

```python
test_df = data2[data2['user_idx'] == test_user_idx]
for i, sr in enumerate(test_df.itertuples()):
    if i > 20:
        break
    print(f'{sr.rating} -> {pred_target_user[sr.book_idx]:.1f}')
```

```python
from sklearn.metrics import mean_squared_error
import math

indices = test_df['book_idx'].to_numpy()
MSE = mean_squared_error(pred_target_user[indices], test_df['rating'].to_list())

RMSE = math.sqrt(MSE)
RMSE
```
