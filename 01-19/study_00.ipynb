```python
import requests
from bs4 import BeautifulSoup as bs

res = requests.get("http://v.media.daum.net/v/20170615203441266")
soup = bs(res.content, "html.parser")
```

```python
soup.find("title")

title = soup.select("title")[0]
print(title)
print(title.get_text())
```

```python
title = soup.select("html head title")[0]
print(title.get_text())
```

```python
import re

res = requests.get('http://media.daum.net/economic/')

soup = bs(res.content, 'html.parser')

# a태그이면서 href 속성을 갖는 경우 탐색, 리스트 타입으로 links 변수에 저장됨
links = soup.select('a[href]')
   
for link in links:
    # print (link) # <a class="link_services link_services2" href="http://sports.media.daum.net/sports">스포츠</a>
    # print (link['href']) # http://sports.media.daum.net/sports
    if re.search('http://\w+', link['href']):  # http:// 문자열 이후에 숫자 또는 문자[a-zA-Z0-9_]가 한 개 이상 있는 데이터와 매치됨 
        print (link['href'])
```

### **API 사용 연습**
* CAR API

```python
from bs4 import BeautifulSoup as bs

url = "https://www.1000islandstoyota.com/new-arrivals/all-toyota-models"
response = requests.get(url)
soup = bs(response.content, 'html.parser')
soup
```

```python
name_toyota = soup.find_all("strong")
type(name_toyota)
```

```python
name_toyota_list = [element.text.strip().split(" ")[1] for element in name_toyota]
name_toyota_list
```

```python
import requests
import json

API_KEY = "VDZiHFgh99CTe4739uVUkQ==29kTOSGJ4vdxlNMj"

all_data = []  # Dictionary to store all data

for name in name_toyota_list:
    api_url = 'https://api.api-ninjas.com/v1/cars?model={}'.format(name)

    response = requests.get(api_url, headers={'X-Api-Key': API_KEY})

    if response.status_code == requests.codes.ok:
        json_data = response.json()
        all_data.extend(json_data)  # Concatenate the JSON data directly
        print(f"Data for {name} added to the dictionary.")
    else:
        print(f"Error for {name}:", response.status_code, response.text)

# Save all data to a single JSON file
with open('all_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(all_data, json_file, ensure_ascii=False, indent=4)

print("All data saved to 'all_data.json'")
```
